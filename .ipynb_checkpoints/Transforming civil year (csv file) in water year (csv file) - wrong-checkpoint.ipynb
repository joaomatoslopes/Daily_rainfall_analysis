```python
import pandas as pd
import numpy as np
import time
import datetime as dt

# Importing coordinates
local = pd.read_csv(r"C:\Users\joaoa\Desktop\Doutoramento\Download ERA5\local.cvs")
local
```

```python
# Creating columns
col_list = ["time"]

for row in range(0, len(local)):
    col_list.append('P' + (4-len(str(local.iloc[row,0])))*'0' + str(local.iloc[row,0]))

    
# Creating dataframe
Y = range(1950, 2023) #2023
P = range(0, 1012) #1012

for year in Y:
    i = time.time()
    y1 = pd.read_csv(r"D:\Precipitation\by_civil_year\era5land_" + str(year) + ".csv", compression = "zip")
    y2 = pd.read_csv(r"D:\Precipitation\by_civil_year\era5land_" + str(year+1) + ".csv", compression = "zip")
    
    df = pd.DataFrame(index = range(0, 8760), columns = col_list)
    df.iloc[:2208, 0] = y1.iloc[-2208:, 3]
    df.iloc[-6552:, 0] = y2.iloc[:6552, 3]
    
    for point in P:
        df.iloc[:2208, point+1] = y1.loc[y1["point"] == local.iloc[point, 0]].iloc[-2208:, 4]
        df.iloc[-6552:, point+1] = y2.loc[y2["point"] == local.iloc[point, 0]].iloc[:6552, 4]
        
    df.to_csv(r"D:\Precipitation\by_water_year\era5land_precipitation_" + str(year) + "_" + str(year+1) + ".csv", compression = "zip", index = False)
    print("Water year " + str(year) + "_" + str(year+1) + " downloaded. Time: " + str(int(time.time()-i)) + " seconds.")
df
```
