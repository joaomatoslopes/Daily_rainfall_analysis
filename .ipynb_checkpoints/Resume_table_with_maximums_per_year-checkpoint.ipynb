```python
import pandas as pd
import numpy as np
```

```python
df_col = pd.read_csv(r"D:\Precipitation\water_year\era5land_total_precipitation_1950_1951_(mm).csv", compression = "zip")
resume = pd.DataFrame(columns = ["year", "type_of_period", "duration"] + list(df_col.columns[1:]))
resume
```

```python
Y = range(1950, 1952) #1950-2023
H = [1,2,4,6,8,12,24,48,72,96,120]
i = -1
    
for year in Y:
    df_year = pd.read_csv(r"D:\Precipitation\water_year\era5land_total_precipitation_" + str(year) + "_" + str(year+1) + "_(mm).csv", compression = "zip")
    
    for hours in H:
        i += 1
        df_roll = df_year.iloc[:, 1:].rolling(hours).sum()
        content = [str(year) + "_" + str(year+1), "no restriction", str(hours) + "h"] + list(df_roll.max())
        resume.loc[i, :] = content

resume
```
